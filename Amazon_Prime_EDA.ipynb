# **Project Summary**
This project focuses on performing Exploratory Data Analysis (EDA) on the Amazon Prime Video dataset, which contains information about movies and TV shows available on the platform.

This Project mainly includes 2 data sets

1 . title.csv (with the following attributes)

* id
* title
* show type
* description
* release year
* age-certification
* runtime
* genres
* production countries
* Seasons
* imdb-id
* imbd-scores
* imbd-votes
* tmbd-popularity
* tmbd-score



2 . Credits.csv (with the following attributes)

* person_id
* id(title id)
* name
* character name


### 1. Objective
The primary objective of this project is to analyze the Amazon Prime Video dataset to understand patterns, trends, and relationships within the data. The goal is to extract meaningful insights that can support data-driven decision-making.

### 2. Data Preparation
The analysis begins with data understanding and cleaning. This includes handling missing values, removing duplicates, and correcting data types. The titles and credits datasets are merged using a common identifier, while carefully managing duplication caused by multiple actors per title.

### 3. Exploratory Data Analysis
The EDA is performed in three stages:
- **Univariate Analysis:** To study distributions of individual features such as IMDb scores, runtime, and release year.
- **Bivariate Analysis:** To explore relationships between variables like ratings vs popularity and runtime vs scores.
- **Multivariate Analysis:** To identify deeper patterns involving multiple variables such as trends over time and content type interactions.

### 4. Contributor & Regional Analysis
The project also analyzes actors and directors to identify frequently appearing contributors. Additionally, production country analysis helps understand global content distribution trends.

### 5. Insights & Impact
The analysis reveals key insights such as dominant genres, content growth trends, and factors affecting ratings and popularity. These insights can help improve content strategy, user engagement, and business decisions.

### 6. Conclusion
Overall, this project transforms raw data into actionable insights and provides a strong foundation for further machine learning and predictive modeling.
