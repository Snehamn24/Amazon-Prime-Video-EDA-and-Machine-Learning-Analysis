# ***3 . Data Wrangling***

Data wrangling (or data munging) is the process of cleaning, structuring, and enriching raw, complex data into a usable format for analysis, machine learning, or visualization.

```python
# 1 . Removing the Duplicates
title.drop_duplicates(inplace=True)
credits.drop_duplicates(inplace=True)
```

### Missing Values Analysis (Titles Dataset)

From the analysis of missing values in the titles dataset, the following observations are made:

- **description** has 119 missing values, which is relatively small and can be ignored or filled with a placeholder.
- **age_certification** has 6484 missing values, indicating that a large portion of content does not have age ratings available.
- **seasons** has 8511 missing values, which is expected since this column is only applicable to TV shows and not movies.
- **imdb_id** has 667 missing values, suggesting incomplete linkage with IMDb records.
- **imdb_score** and **imdb_votes** have over 1000 missing values, which may impact rating-based analysis.
- **tmdb_popularity** and **tmdb_score** also contain missing values, though comparatively fewer.

### Key Insight

Missing values are concentrated in specific columns rather than being randomly distributed. Some missing values are logical (e.g., seasons for movies), while others indicate incomplete data.

### Conclusion

Appropriate handling strategies such as imputation, replacement, or column removal are required to ensure data quality before analysis.

```python
# Fill description with placeholder
title['description']=title['description'].fillna("No description available")

# Fill age certification
title['age_certification']=title['age_certification'].fillna("Unknown")

# Drop seasons (not useful for all rows)
#title.drop(columns=['seasons'], inplace=True)


# Fill ratings with median
title['imdb_score']=title['imdb_score'].fillna(title['imdb_score'].median())
title['tmdb_score'] = title['tmdb_score'].fillna(title['tmdb_score'].median())


# Fill votes/popularity with 0
title['imdb_votes'] = title['imdb_votes'].fillna(0)
title['tmdb_popularity'] = title['tmdb_popularity'].fillna(0)
```

```python
#check for the null values
title.isna().sum()
```
